```python
#2. Write a code to count the number of vowels in a string.
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Example usage
input_string = "Hello, World!"
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)
```

```python
#3. Write a code to check if a given string is a palindrome or not.
def is_palindrome(s):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    s = s.replace(" ", "").lower()
    # Check if the string is equal to its reverse
    return s == s[::-1]

# Example usage
input_string = "A man a plan a canal Panama"
if is_palindrome(input_string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
#4. Write a code to check if two given strings are anagrams of each other
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase for a case-insensitive comparison
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Sort the characters of both strings and compare
    return sorted(str1) == sorted(str2)

# Example usage
string1 = "Listen"
string2 = "Silent"

if are_anagrams(string1, string2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

```python
#5. Write a code to find all occurrences of a given substring within another string
def find_occurrences(main_string, substring):
    start = 0
    positions = []
    while True:
        start = main_string.find(substring, start)
        if start == -1:
            break
        positions.append(start)
        start += 1  # Move past the last found substring
    return positions

# Example usage
main_string = "abracadabra"
substring = "abra"

positions = find_occurrences(main_string, substring)
print("Occurrences found at positions:", positions)
```

```python
#6.  Write a code to perform basic string compression using the counts of repeated characters
def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Add the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a string
    compressed_string = ''.join(compressed)

    # Return the original string if compressed string isn't shorter
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage
input_string = "aabcccccaaa"
compressed_string = compress_string(input_string)
print("Compressed String:", compressed_string)
```

```python
#7. Write a code to determine if a string has all unique characters.
def has_unique_characters(s):
    # Using a set to store characters we have seen
    seen_characters = set()

    for char in s:
        if char in seen_characters:
            return False  # Found a duplicate character
        seen_characters.add(char)

    return True  # All characters are unique

# Example usage
input_string = "abcdefg"
if has_unique_characters(input_string):
    print("The string has all unique characters.")
else:
    print("The string does not have all unique characters.")
```

```python
#12.  Implement a code to find the second largest number in a given list of integers.
def find_second_largest(numbers):
    if len(numbers) < 2:
        raise ValueError("List must contain at least two elements")

    # Initialize the largest and second largest to negative infinity
    largest = second_largest = float('-inf')

    for number in numbers:
        if number > largest:
            second_largest = largest
            largest = number
        elif number > second_largest and number < largest:
            second_largest = number

    if second_largest == float('-inf'):
        raise ValueError("There is no second largest number")

    return second_largest

# Example usage
my_list = [10, 20, 4, 45, 99]
second_largest = find_second_largest(my_list)
print("Second largest number:", second_largest)
```

```python
#14. Write a code to reverse a list in-place without using any built-in reverse functions.
def reverse_list_in_place(lst):
    start = 0
    end = len(lst) - 1

    while start < end:
        # Swap the elements at start and end indices
        lst[start], lst[end] = lst[end], lst[start]
        start += 1
        end -= 1

# Example usage
my_list = [1, 2, 3, 4, 5]
reverse_list_in_place(my_list)
print("Reversed List:", my_list)
```

```python
#15. Implement a code to find and remove duplicates from a list while preserving the original order of elements.
def remove_duplicates(lst):
    seen = set()  # To keep track of seen elements
    result = []   # To store elements without duplicates

    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)

    return result

# Example usage
my_list = [1, 2, 3, 2, 4, 1, 5, 6, 4]
no_duplicates = remove_duplicates(my_list)
print("List without duplicates:", no_duplicates)
```

```python
#16. Create a code to check if a given list is sorted (either in ascending or descending order) or not.
def is_sorted(lst):
    if not lst:
        return True  # An empty list is considered sorted

    ascending = descending = True

    for i in range(1, len(lst)):
        if lst[i] < lst[i - 1]:
            ascending = False
        if lst[i] > lst[i - 1]:
            descending = False

    return ascending or descending

# Example usage
my_list = [1, 2, 3, 4, 5]
is_sorted_list = is_sorted(my_list)
print("The list is sorted:", is_sorted_list)

my_list_desc = [5, 4, 3, 2, 1]
is_sorted_list_desc = is_sorted(my_list_desc)
print("The list is sorted:", is_sorted_list_desc)

my_list_unsorted = [1, 3, 2, 4, 5]
is_sorted_list_unsorted = is_sorted(my_list_unsorted)
print("The list is sorted:", is_sorted_list_unsorted)
```

```python
#17. Write a code to merge two sorted lists into a single sorted list.
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = j = 0

    # Traverse both lists and append smaller element to merged_list
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Append any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
merged_list = merge_sorted_lists(sorted_list1, sorted_list2)
print("Merged Sorted List:", merged_list)
```

```python
#18. Implement a code to find the intersection of two given lists.
def intersection_of_lists(list1, list2):
    # Convert lists to sets to find the intersection
    set1 = set(list1)
    set2 = set(list2)

    # Find intersection
    intersection = set1.intersection(set2)

    # Convert the result back to a list
    return list(intersection)

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
intersection = intersection_of_lists(list1, list2)
print("Intersection of the lists:", intersection)
```

```python
#20.  Write a code to shuffle a given list randomly without using any built-in shuffle functions.
import random

def shuffle_list(lst):
    # Create a copy of the list to avoid modifying the original
    shuffled_list = lst[:]
    n = len(shuffled_list)

    # Shuffle the list using the Fisher-Yates algorithm
    for i in range(n - 1, 0, -1):
        j = random.randint(0, i)
        # Swap elements at index i and j
        shuffled_list[i], shuffled_list[j] = shuffled_list[j], shuffled_list[i]

    return shuffled_list

# Example usage
my_list = [1, 2, 3, 4, 5]
shuffled = shuffle_list(my_list)
print("Shuffled List:", shuffled)
```

```python
#25.  Create a code that takes a tuple and two integers as input. The function should return a new tuple containing elements from the original tuple within the specified range of indices.
def slice_tuple(tup, start, end):
    # Ensure start and end are within the valid range
    if start < 0:
        start = 0
    if end > len(tup):
        end = len(tup)

    # Return the sliced tuple
    return tup[start:end]

# Example usage
original_tuple = (10, 20, 30, 40, 50, 60, 70)
start_index = 2
end_index = 5
sliced_tuple = slice_tuple(original_tuple, start_index, end_index)
print("Sliced Tuple:", sliced_tuple)
```

```python
#31. Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list.
def word_frequencies(word_list):
    # Initialize an empty dictionary to store word frequencies
    frequency_dict = {}

    # Iterate through each word in the list
    for word in word_list:
        # If the word is already in the dictionary, increment its count
        if word in frequency_dict:
            frequency_dict[word] += 1
        # Otherwise, add the word to the dictionary with a count of 1
        else:
            frequency_dict[word] = 1

    # Return the dictionary with word frequencies
    return frequency_dict

# Example usage
input_words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
frequencies = word_frequencies(input_words)
print("Word Frequencies:", frequencies)
```

```python
#32.  Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together.
def merge_dictionaries(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = dict1.copy()  # Start with a copy of the first dictionary

    # Iterate through the second dictionary
    for key, value in dict2.items():
        # If the key is already in the merged dictionary, add the values
        if key in merged_dict:
            merged_dict[key] += value
        # Otherwise, add the key-value pair to the merged dictionary
        else:
            merged_dict[key] = value

    return merged_dict

# Example usage
dict1 = {'apple': 3, 'banana': 2, 'cherry': 5}
dict2 = {'banana': 4, 'cherry': 1, 'date': 7}
merged_dict = merge_dictionaries(dict1, dict2)
print("Merged Dictionary:", merged_dict)
```

```python
#33.  Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, thefunction should return Non.
def get_nested_value(nested_dict, keys):
    # Start with the full nested dictionary
    current_value = nested_dict

    # Iterate through the list of keys
    for key in keys:
        # Check if the current key exists in the current level of the dictionary
        if isinstance(current_value, dict) and key in current_value:
            current_value = current_value[key]  # Move to the next level
        else:
            return None  # If the key is not found, return None

    return current_value  # Return the value found

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'x': {
        'y': 100
    }
}
keys = ['a', 'b', 'c']
result = get_nested_value(nested_dict, keys)
print("Value:", result)  # Output should be 42

# Example with a missing key
keys = ['a', 'b', 'z']
result = get_nested_value(nested_dict, keys)
print("Value:", result)  # Output should be None
```

```python
#34. Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.

def sort_dict_by_value(input_dict, descending=False):
    # Sort the dictionary by its values
    sorted_items = sorted(input_dict.items(), key=lambda item: item[1], reverse=descending)
    # Convert the sorted items back to a dictionary
    sorted_dict = dict(sorted_items)
    return sorted_dict

# Example usage
input_dict = {'apple': 3, 'banana': 2, 'cherry': 5, 'date': 1}

# Sort in ascending order
sorted_dict_asc = sort_dict_by_value(input_dict, descending=False)
print("Sorted Dictionary (Ascending):", sorted_dict_asc)

# Sort in descending order
sorted_dict_desc = sort_dict_by_value(input_dict, descending=True)
print("Sorted Dictionary (Descending):", sorted_dict_desc)
```

```python
#35.  Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

def invert_dictionary(input_dict):
    # Initialize an empty dictionary to store the inverted dictionary
    inverted_dict = {}

    # Iterate through each key-value pair in the original dictionary
    for key, value in input_dict.items():
        # If the value is already a key in the inverted dictionary
        if value in inverted_dict:
            inverted_dict[value].append(key)  # Append the current key to the list
        else:
            inverted_dict[value] = [key]  # Create a new list with the current key

    return inverted_dict

# Example usage
input_dict = {'a': 1, 'b': 2, 'c': 1, 'd': 3}

inverted_dict = invert_dictionary(input_dict)
print("Inverted Dictionary:", inverted_dict)
```
